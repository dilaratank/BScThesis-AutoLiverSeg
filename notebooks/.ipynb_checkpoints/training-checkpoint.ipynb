### Dice similarity:
"This coefficient measures the similarity between sets X and Y. If the two sets are identical (i.e. they contain the same elements), the coefficient is equal to 1.0, while if X and Y have no elements in common, it is equal to 0.0. Otherwise it is somewhere in between."

```python
# function to plot image, mask and prediction together 
def vis_pred(image, mask, pred):

    with torch.no_grad():
        fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (12,6))

        ax1.imshow(image.squeeze(0), cmap='gray')
        ax1.set_title('Image')

        ax2.imshow(mask.squeeze(0), cmap='gray')
        ax2.set_title('Ground Truth')

        ax3.imshow(pred.squeeze(0).squeeze(0), cmap='gray')
        ax3.set_title('Prediction')

        plt.plot()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Apr 21 16:18:27 2021

@author: dtank
"""

import sys  
sys.path.append('.')
sys.path.append('../src/utils')

import mt_sourcecode as mt_adap
from data import *
from medicaltorch import transforms as mt_transforms
from medicaltorch import datasets as mt_datasets
from medicaltorch import models as mt_models
from medicaltorch import losses as mt_losses
import torch
from torch.utils.data import DataLoader
import torch.optim as optim
from tqdm import tqdm
import IO as io
import warnings
warnings.filterwarnings('ignore')

def train():
    # Make filenames
    data_path = '/home/dtank/lood_storage/divi/Projects/anchor/analysis/'
    dirlist = io.main(data_path)
    
    filename_pairs = make_filename_pairs(dirlist)
    train_filename_pairs = filename_pairs[2:5] # use only small section for tryout
    val_filename_pairs = filename_pairs[:1]
    
    # Device
#     device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    device = 'cpu'
    
    # Network
    network = mt_models.Unet()
    network.to(device)

    # Dataset and Dataloader
    train_dataset = mt_adap.MRI2DSegmentationDataset(train_filename_pairs,transform=mt_transforms.ToTensor()) 
    train_dataloader = DataLoader(train_dataset, batch_size=16,collate_fn=mt_datasets.mt_collate)
    
    val_dataset = mt_adap.MRI2DSegmentationDataset(val_filename_pairs,transform=mt_transforms.ToTensor()) 
    val_dataloader = DataLoader(train_dataset, batch_size=16,collate_fn=mt_datasets.mt_collate)
    
    # Optimizer
    optimizer = optim.Adam(network.parameters(), lr=0.01) # optimizer has networks weights (parameters) --> can  update

    # Training Loop    
    total_loss = 0
    
    print('train')
    for batch in tqdm(train_dataloader, position=0, leave=True):
        images, labels = batch["input"], batch["gt"]
        images, labels = images.to(device), labels.to(device)
        
        network.train()
        preds = network(images) # pass batch
        
        loss = mt_losses.dice_loss(preds, labels) # calculate loss
        dsc = dice_score(preds, labels)
        print('loss:', loss.item())
    
        optimizer.zero_grad() # pytorch accumulates gradients, we don't want that
        loss.backward() # calculate gradients
        optimizer.step() # update weights 
        
        total_loss += loss.item()
    
    # Eval loop
    total_score = 0
    
    print('eval')
    with torch.no_grad():
        for batch in tqdm(val_dataloader, position=0, leave=True):
            images, labels = batch['input'], batch['gt']
            images, labels = images.to(device), labels.to(device)
            
            network.eval()
            pred = network(images)
            loss = mt_losses.dice_loss(pred, labels)
            dsc = -1 * mt_losses.dice_loss(pred, labels)
            print('loss:', loss.item())
            print('dice score:', dsc.item())
            total_score += dsc.item()
            
            
        
    print('epoch:', 0, 'loss:', total_loss/len(train_dataloader), 'score:', total_score/len(val_dataloader))
```

```python
def sub_filename_pairs(filename_pairs, train_index, test_index):
    train = []
    test = []
    for index in train_index:
        train.append(filename_pairs[index])
    for index in test_index:
        test.append(filename_pairs[index])
    return train, test

def reset_weights(m):
    for layer in m.children():
        if hasattr(layer, 'reset_parameters'):
            layer.reset_parameters()
            
def training_loop(network, train_dataloader, num_epochs, device, fold):
    for epoch in range (0, num_epochs):
        
        # print epoch
        print(f'Starting epoch {epoch+1}')
        
        # set current loss value
        current_loss = 0
        
        # iterate over dataloader for training data
        for i, batch in enumerate(tqdm.tqdm(train_dataloader, position=0, leave=True)):
            
            # get inputs
            images, labels = batch["input"], batch["gt"]
            images, labels = images.to(device), labels.to(device)

            # zero the gradients
            optimizer.zero_grad() 
            
            # perform forward pass
            network.train()
            preds = network(images) 
            
            # compute loss
            loss = mt_losses.dice_loss(preds, labels) 
            
            # perform backward pass
            loss.backward() 
            
            # perform optimization 
            optimizer.step() 
            
            # print statistics 
            # TODO: save to tensorboard?
            current_loss += loss.item()
            if i%10 == 9:
                print('Loss after mini-batch %5d: %.3f' % (i+1, current_loss / 10))
                current_loss = 0
                
        # process complete, save model
        # TODO: save to tensorboard?
        print('Training process finished, save model')
        save_path = f'./model-fold-{fold}.pth'
        torch.save(network.state_dict(), save_path)
        
def testing_loop(network, test_dataloader, results, device, fold):
    print('Start testing')

    # evaluation for current fold
    with torch.no_grad():
        for batch in tqdm.tqdm(test_dataloader, position=0, leave=True):

            # total score
            total_score = 0

            # get inputs
            images, labels = batch['input'], batch['gt']
            images, labels = images.to(device), labels.to(device)

            # generate outpuds 
            network.eval()
            pred = network(images)

            # calculate dice coefficient 
            dsc = -1 * mt_losses.dice_loss(pred, labels)
            total_score += dsc.item()

        # print dice coefficient score 
        print(f'Dice score for fold {fold}:', total_score/len(batch))
        print('------------------------------')
        results[fold] = total_score/len(batch)
    
    return results
```

```python
# configuration options
k_folds = 5
num_epochs = 1

# device
# device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device = 'cpu'

# define transformations
train_transforms = torchvision.transforms.Compose([
        mt_transforms.RandomRotation(10), # random rotation
        mt_transforms.RandomAffine(0, translate=(0.05, 0.05)), # shift
        mt_transforms.RandomAffine(0, shear=(-5, 5)), # shear
        mt_transforms.AdditiveGaussianNoise(std=0.3), # noise 
        mt_transforms.ToTensor(),
    ])
test_transforms = torchvision.transforms.Compose([mt_transforms.ToTensor()])

# fold results
results = {}

# set fixed random number seed so that pseudo random number  
# initializers will be initialized using the same initialization token.
torch.manual_seed(42)

# define cross validatior
kfold = KFold(n_splits=k_folds, shuffle=True)

# start print
print('------------------------------')

for fold, (train_index, test_index) in enumerate(kfold.split(filename_pairs)):
    print(f'FOLD {fold}')
    print('------------------------------')
    
    # get train and test filenames
    train_files, test_files = sub_filename_pairs(filename_pairs, train_index, test_index)
    
    # define datasets
    train_dataset = mt_adap.MRI2DSegmentationDataset(train_files,transform=train_transforms, p=0.8)
    test_dataset = mt_adap.MRI2DSegmentationDataset(test_files,transform=test_transforms)
    
    # define dataloaders
    train_dataloader = DataLoader(train_dataset, batch_size=16,collate_fn=mt_datasets.mt_collate)
    test_dataloader = DataLoader(test_dataset, batch_size=16,collate_fn=mt_datasets.mt_collate)
    
    # init the neural network 
    network = mt_models.Unet()
    network.apply(reset_weights)
    
    # initialize optimizer
    optimizer = optim.Adam(network.parameters(), lr=0.01)
    
    # train
    training_loop(network, train_dataloader, num_epochs, device, fold)
    
    # test
    results = testing_loop(network, test_dataloader, results, device, fold)
    
# print fold results
print(f'K-FOLD CROSS VALIDATION RESULTS FOR {k_folds} FOLDS')  
print('------------------------------')
sum = 0
for key, value in results.items():
    print(f'Fold {key}: {value}')
    sum += value
print(f'Average:', sum/len(results.items()))
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Apr 29 15:43:25 2021
training loop adapted from https://www.machinecurve.com/index.php/2021/02/03/how-to-use-k-fold-cross-validation-with-pytorch/

@author: dtank
"""

# imports
from sklearn.model_selection import KFold
import sys  
sys.path.append('.')
sys.path.append('../src/utils')
import mt_sourcecode as mt_adap
from utils import *
from medicaltorch import transforms as mt_transforms
from medicaltorch import datasets as mt_datasets
from medicaltorch import models as mt_models
from medicaltorch import losses as mt_losses
import yaml
import numpy as np
import torch
import torchvision
from torch.utils.tensorboard import SummaryWriter
from torch.utils.data import DataLoader
import torch.optim as optim
from tqdm import tqdm
import IO as io
import warnings
warnings.filterwarnings('ignore')

# helper functions
def slice_filtering(slice_pair):
    if len(np.unique(slice_pair['gt'])) == 1:
        return False
    else:
        return True
    
def sub_filename_pairs(filename_pairs, train_index, test_index):
    train = []
    test = []
    for index in train_index:
        train.append(filename_pairs[index])
    for index in test_index:
        test.append(filename_pairs[index])
    return train, test

def reset_weights(m):
    for layer in m.children():
        if hasattr(layer, 'reset_parameters'):
            layer.reset_parameters()
            
def training_loop(network, train_dataloader, num_epochs, device, optimizer, fold):
    for epoch in range (0, num_epochs):
        
        # print epoch
        print(f'Starting epoch {epoch+1}')
        
        # set current loss value
        current_loss = 0
        
        # iterate over dataloader for training data
        for i, batch in enumerate(tqdm(train_dataloader, position=0, leave=True)):
            
            # get inputs
            images, labels = batch["input"], batch["gt"]
            images, labels = images.to(device), labels.to(device)

            # zero the gradients
            optimizer.zero_grad() 
            
            # perform forward pass
            network.train()
            preds = network(images) 
            
            # compute loss
            if config['train_params']['loss'] == 'dice_loss':
                loss = mt_losses.dice_loss(preds, labels)
                # tb.add_scalar('loss', loss.item())
            elif config['train_params']['loss'] == 'BCE':
                lossf = torch.nn.BCELoss()
                loss = lossf(preds, labels)
            
            # perform backward pass
            loss.backward() 
            
            # perform optimization 
            optimizer.step() 
            
            current_loss += loss.item()
            
                
        # process complete, save model
        print(f'Training process epoch {epoch} finished, save model')
        print('Current loss:', current_loss/len(train_dataloader))
        tb.add_scalar('Train Loss', current_loss/len(train_dataloader), global_step=epoch)
        # tb.flush()
        save_path = f'./model-fold-{fold}.pth'
        torch.save(network.state_dict(), save_path)
        
def testing_loop(network, test_dataloader, results, device, fold):
    print('Start testing')
    
    # total score
    total_score, total = 0, 0

    # evaluation for current fold
    with torch.no_grad():
        for batch in tqdm(test_dataloader, position=0, leave=True):

            # get inputs
            images, labels = batch['input'], batch['gt']
            images, labels = images.to(device), labels.to(device)

            # generate outpuds 
            network.eval()
            pred = network(images)

            # calculate dice coefficient 
            dsc = -1 * mt_losses.dice_loss(pred, labels)
            print('dice score:', dsc.item())
            total_score += dsc.item()
            total += 1

        # print dice coefficient score 
        print(f'Dice score for fold {fold}:', total_score/total)
        tb.add_scalar('Dice Score Test', total_score/total)
        print('total score:', total_score)
        print('total preds:', total)
        print('------------------------------')
        results[fold] = total_score/total
    
    return results

def train(config):
    # get data
    dirlist = io.main(config['dataset']['data_directory'])
    filename_pairs = make_filename_pairs(dirlist)
    
    # device
    device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
    
    # define transformations
    train_transforms = torchvision.transforms.Compose([
            mt_transforms.RandomRotation(config['train_dataloader']['data_aug']['rot_degree']), # random rotation
            # mt_transforms.RandomAffine(0, translate=config['train_dataloader']['data_aug']['transl_range']), # shift
            mt_transforms.RandomAffine(0, shear=config['train_dataloader']['data_aug']['shear_range']), # shear
            # mt_transforms.AdditiveGaussianNoise(std=config['train_dataloader']['data_aug']['gauss_params']['std']), # noise 
            mt_adap.RangeNorm(),
            mt_adap.HistogramClipping(),
            mt_transforms.ToTensor(),
        ])
    test_transforms = torchvision.transforms.Compose([
                                                      mt_adap.RangeNorm(),
                                                      mt_adap.HistogramClipping(),
                                                      mt_transforms.ToTensor()])
    
    # fold results
    results = {}
    
    # set fixed random number seed so that pseudo random number  
    # initializers will be initialized using the same initialization token.    
    # define cross validatior
    kfold = KFold(n_splits=config['model_params']['k_folds'], 
                  shuffle=True, random_state=config['model_params']['random_state'])
    
    # start print
    print('------------------------------')
    
    for fold, (train_index, test_index) in enumerate(kfold.split(filename_pairs)):
        print(f'FOLD {fold}')
        print('------------------------------')
        
        # get train and test filenames
        train_files, test_files = sub_filename_pairs(filename_pairs, train_index, test_index)
        
        # define datasets
        train_dataset = mt_adap.MRI2DSegmentationDataset(train_files,
                                                         transform=train_transforms, 
                                                         slice_filter_fn=slice_filtering,
                                                         p=0.8)
        
        test_dataset = mt_adap.MRI2DSegmentationDataset(test_files,
                                                        transform=test_transforms,
                                                        slice_filter_fn=slice_filtering)
        
        # define dataloaders
        train_dataloader = DataLoader(train_dataset, 
                                      batch_size=config['train_dataloader']['batch_size'],
                                      shuffle=config['train_dataloader']['shuffle'],
                                      collate_fn=mt_datasets.mt_collate)
        
        test_dataloader = DataLoader(test_dataset, 
                                     batch_size=config['test_dataloader']['batch_size'],
                                     collate_fn=mt_datasets.mt_collate)
        
        # init the neural network
        if config['model_params']['architecture'] == 'unet':
            network = mt_models.Unet()
        elif config['model_params']['architecture'] == 'segnet':
            pass #TODO: add segnet model
        network.apply(reset_weights)
        network.to(device)
        
        # initialize optimizer
        if config['train_params']['optimizer'] == 'adam':
            optimizer = optim.Adam(network.parameters(), lr=config['train_params']['lr'])
        
        # train
        training_loop(network, train_dataloader, 
                      config['train_params']['num_epochs'], 
                      device, optimizer, fold)
        
        # test
        results = testing_loop(network, test_dataloader, 
                               results, device, fold)
        
        break 
        
    # print fold results
    k_folds = config['model_params']['k_folds']
    print(f'K-FOLD CROSS VALIDATION RESULTS FOR {k_folds} FOLDS')  
    print('------------------------------')
    sum_res = 0
    for key, value in results.items():
        print(f'Fold {key}: {value}')
        # tb.add_text(f'Fold {key}: {value}')
        sum_res += value
    print('Average:', sum_res/len(results.items()))
    
if __name__ == "__main__":
    
    CONFIG_PATH = '/home/dtank/scratch/AutoLiverSeg/src/config.yaml'
    config = load_config(CONFIG_PATH)
    
    # tensorboard logs setup
    comment = 'Slice filtering, range norm, histclip(1,99), randomrot, shear'
    tb = SummaryWriter(comment=comment)
    
    train(config)
    
    tb.close()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Apr 29 15:43:25 2021
training loop adapted from https://www.machinecurve.com/index.php/2021/02/03/how-to-use-k-fold-cross-validation-with-pytorch/

@author: dtank
"""

# imports
from sklearn.model_selection import KFold
import sys  
sys.path.append('.')
sys.path.append('../src/utils')
# import mt_sourcecode as mt_adap
# import mt_sourcecode_avgimg as mt_adap # FOR TRAINING WITH AVERAGE IMAGE
import mt_sourcecode_multinput as mt_adap # FOR TRAINING WITH MULTI INPUT
from utils import *
from medicaltorch import transforms as mt_transforms
from medicaltorch import datasets as mt_datasets
from medicaltorch import models as mt_models
from medicaltorch import losses as mt_losses
import yaml
import numpy as np
from sklearn.model_selection import train_test_split
import torch
import torchvision
from torch.utils.tensorboard import SummaryWriter
from torch.utils.data import DataLoader
import torch.optim as optim
from tqdm import tqdm
import IO as io
import warnings
warnings.filterwarnings('ignore')

# helper functions
def slice_filtering(slice_pair):
    if len(np.unique(slice_pair['gt'])) == 1:
        return False
    else:
        return True
    
def sub_filename_pairs(filename_pairs, train_index, test_index):
    train = []
    test = []
    for index in train_index:
        train.append(filename_pairs[index])
    for index in test_index:
        test.append(filename_pairs[index])
    return train, test

def reset_weights(m):
    for layer in m.children():
        if hasattr(layer, 'reset_parameters'):
            layer.reset_parameters()
            
def training_loop(network, train_dataloader, val_dataloader, num_epochs, device, optimizer, fold):
    early_stopping_counter = 0
    best_loss = 0
    for epoch in range (0, num_epochs):
        
        # print epoch
        print(f'Starting epoch {epoch+1}')
        
        # set current loss value
        current_loss = 0
        
        # iterate over dataloader for training data
        for i, batch in enumerate(tqdm(train_dataloader, position=0, leave=True)):
            
            # get inputs
            images, labels = batch["input"], batch["gt"]
            images, labels = images.to(device), labels.to(device)

            # zero the gradients
            optimizer.zero_grad() 
            
            # perform forward pass
            network.train()
            preds = network(images) 
            
            # compute loss
            if config['train_params']['loss'] == 'dice_loss':
                loss = mt_losses.dice_loss(preds, labels)
                # tb.add_scalar('loss', loss.item())
            elif config['train_params']['loss'] == 'BCE':
                lossf = torch.nn.BCELoss()
                loss = lossf(preds, labels)
            
            # perform backward pass
            loss.backward() 
            
            # perform optimization 
            optimizer.step() 
            
            current_loss += loss.item()
            
                
        # process complete, print statistics, validate and save model
        print(f'Training process epoch {epoch} finished')
        print('Current loss:', current_loss/len(train_dataloader))
        tb.add_scalar(f'Train Loss Fold {fold}', current_loss/len(train_dataloader), global_step=epoch)
        # tb.flush()
       
        # validation
        print(f'Validating epoch {epoch}')
        val_loss = 0
        
        with torch.no_grad():
            for batch in tqdm(val_dataloader, position=0, leave=True):
    
                # get inputs
                images, labels = batch['input'], batch['gt']
                images, labels = images.to(device), labels.to(device)
    
                # generate outpuds 
                network.eval()
                preds = network(images)
                
                loss = mt_losses.dice_loss(preds, labels)
                val_loss += loss.item()
        
        mean_val_loss = val_loss/len(val_dataloader)
        
        print('mean validation loss', mean_val_loss)
        print('best loss', best_loss)
        
        if epoch % 5 == 0:
            if mean_val_loss < best_loss:
                best_loss = mean_val_loss
                save_path = f'./model-fold-{fold}.pth'
                torch.save(network.state_dict(), save_path)
            else:
                early_stopping_counter += 1
            
        tb.add_scalar(f'Validation Loss Fold {fold}', mean_val_loss, global_step=epoch)
        
        if early_stopping_counter == 25:
            # save_path = f'./model-fold-{fold}.pth'
            # torch.save(network.state_dict(), save_path)
            print('Early stopping, Model saved')
            break
        
        print(f'Early stopping counter: {early_stopping_counter}')
            
def testing_loop(network, test_dataloader, results, device, fold):
    print('Start testing')
    
    # total score
    total_score, total = 0, 0

    # evaluation for current fold
    with torch.no_grad():
        for batch in tqdm(test_dataloader, position=0, leave=True):

            # get inputs
            images, labels = batch['input'], batch['gt']
            images, labels = images.to(device), labels.to(device)

            # generate outpuds 
            network.eval()
            pred = network(images)

            # calculate dice coefficient 
            dsc = -1 * mt_losses.dice_loss(pred, labels)
            print('dice score:', dsc.item())
            total_score += dsc.item()
            total += 1

        # print dice coefficient score 
        print(f'Dice score for fold {fold}:', total_score/total)
        tb.add_scalar(f'Dice Score Test Fold {fold}', total_score/total)
        print('total score:', total_score)
        print('total preds:', total)
        print('------------------------------')
        results[fold] = total_score/total
    
    return results

def train(config):
    # get data
    dirlist = io.main(config['dataset']['data_directory'])
    filename_pairs = make_filename_pairs(dirlist)
    
    # device
    device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
    
    # define transformations
    train_transforms = torchvision.transforms.Compose([
            mt_transforms.RandomRotation(config['train_dataloader']['data_aug']['rot_degree']), # random rotation
            mt_transforms.RandomAffine(0, translate=config['train_dataloader']['data_aug']['transl_range']), # shift
            mt_transforms.RandomAffine(0, shear=config['train_dataloader']['data_aug']['shear_range']), # shear
            # mt_transforms.AdditiveGaussianNoise(mean=config['train_dataloader']['data_aug']['gauss_params']['mean'] ,std=config['train_dataloader']['data_aug']['gauss_params']['std']), # noise 
            mt_adap.HistogramClipping(),
            mt_adap.RangeNorm(),
            mt_transforms.ToTensor(),
        ])
    test_val_transforms = torchvision.transforms.Compose([
                                                      mt_adap.HistogramClipping(),
                                                      mt_adap.RangeNorm(),
                                                      mt_transforms.ToTensor()])
    
    # fold results
    results = {}
    
    # set fixed random number seed so that pseudo random number  
    # initializers will be initialized using the same initialization token.    
    # define cross validatior
    kfold = KFold(n_splits=config['model_params']['k_folds'], 
                  shuffle=True, random_state=config['model_params']['random_state'])
    
    # start print
    print('------------------------------')
    
    for fold, (train_index, test_index) in enumerate(kfold.split(filename_pairs)):
        print(f'FOLD {fold}')
        print('------------------------------')
        
        # get train and test filenames
        train_files, test_files = sub_filename_pairs(filename_pairs, train_index, test_index)
        
        # get train and validation filenames
        train_files, val_files = train_test_split(train_files, test_size=0.1, 
                                                  random_state=config['model_params']['random_state'])
        
        # define datasets
        train_dataset = mt_adap.MRI2DSegmentationDataset(train_files,
                                                         transform=train_transforms, 
                                                         slice_filter_fn=slice_filtering,
                                                         p=0.8)
        
        val_dataset = mt_adap.MRI2DSegmentationDataset(val_files, 
                                                       transform=test_val_transforms, 
                                                       slice_filter_fn=slice_filtering)
        
        test_dataset = mt_adap.MRI2DSegmentationDataset(test_files,
                                                        transform=test_val_transforms,
                                                        slice_filter_fn=slice_filtering)
        print('datasets completed')
        
        # define dataloaders
        train_dataloader = DataLoader(train_dataset, 
                                      batch_size=config['train_dataloader']['batch_size'],
                                      shuffle=config['train_dataloader']['shuffle'],
                                      collate_fn=mt_datasets.mt_collate)
        
        val_dataloader = DataLoader(val_dataset, 
                                    batch_size=config['test_dataloader']['batch_size'],
                                    collate_fn=mt_datasets.mt_collate)
        
        test_dataloader = DataLoader(test_dataset, 
                                     batch_size=config['test_dataloader']['batch_size'],
                                     collate_fn=mt_datasets.mt_collate)
        
        # init the neural network
        if config['model_params']['architecture'] == 'unet':
            network = mt_models.Unet()
        elif config['model_params']['architecture'] == 'segnet':
            pass #TODO: add segnet model
        network.apply(reset_weights)
        network.to(device)
        
        # initialize optimizer
        if config['train_params']['optimizer'] == 'adam':
            optimizer = optim.Adam(network.parameters(), lr=config['train_params']['lr'])
        
        # train
        training_loop(network, train_dataloader, val_dataloader,
                      config['train_params']['num_epochs'], 
                      device, optimizer, fold)

        # get best model        
        PATH = f'./model-fold-{fold}.pth'
        testnetwork = mt_models.Unet()
        testnetwork.load_state_dict(torch.load(PATH))
        testnetwork.to(device)
        # test
        results = testing_loop(testnetwork, test_dataloader, 
                               results, device, fold)
        
        break 
        
    # print fold results
    k_folds = config['model_params']['k_folds']
    print(f'K-FOLD CROSS VALIDATION RESULTS FOR {k_folds} FOLDS')  
    print('------------------------------')
    sum_res = 0
    for key, value in results.items():
        print(f'Fold {key}: {value}')
        # tb.add_text(f'Fold {key}: {value}')
        sum_res += value
    print('Average:', sum_res/len(results.items()))
    
if __name__ == "__main__":
    
    CONFIG_PATH = '/home/dtank/scratch/AutoLiverSeg/src/config.yaml'
    config = load_config(CONFIG_PATH)
    
    # tensorboard logs setup
    comment = 'Train with Average b=0 images'
    tb = SummaryWriter(comment=comment)
    
    train(config)
    
    tb.close()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Apr 29 15:43:25 2021
training loop adapted from https://www.machinecurve.com/index.php/2021/02/03/how-to-use-k-fold-cross-validation-with-pytorch/

@author: dtank
"""

# imports
from sklearn.model_selection import KFold
import sys  
sys.path.append('.')
sys.path.append('../src/utils')
from datasets import *
from transforms import *
from models import *
from utils import *
from medicaltorch import transforms as mt_transforms
from medicaltorch import datasets as mt_datasets
from medicaltorch import models as mt_models
from medicaltorch import losses as mt_losses
# import yaml
import numpy as np
from sklearn.model_selection import train_test_split
import torch
import torchvision
from torch.utils.tensorboard import SummaryWriter
from torch.utils.data import DataLoader
import torch.optim as optim
from tqdm import tqdm
import IO as io
import warnings
warnings.filterwarnings('ignore')
            
def training_loop(network, train_dataloader, val_dataloader, num_epochs, device, optimizer, fold):
    early_stopping_counter = 0
    best_loss = 0
    for epoch in range (0, num_epochs):
        
        # print epoch
        print(f'Starting epoch {epoch+1}')
        
        # set current loss value
        current_loss = 0
        
        # iterate over dataloader for training data
        for i, batch in enumerate(tqdm(train_dataloader, position=0, leave=True)):
            
            # get inputs
            images, labels = batch["input"], batch["gt"]
            images, labels = images.to(device), labels.to(device)

            # zero the gradients
            optimizer.zero_grad() 
            
            # perform forward pass
            network.train()
            preds = network(images) 
            
            # compute loss
            if config['train_params']['loss'] == 'dice_loss':
                loss = mt_losses.dice_loss(preds, labels)
                # tb.add_scalar('loss', loss.item())
            elif config['train_params']['loss'] == 'BCE':
                lossf = torch.nn.BCELoss()
                loss = lossf(preds, labels)
            
            # perform backward pass
            loss.backward() 
            
            # perform optimization 
            optimizer.step() 
            
            current_loss += loss.item()
            
                
        # process complete, print statistics, validate and save model
        print(f'Training process epoch {epoch} finished')
        print('Current loss:', current_loss/len(train_dataloader))
        tb.add_scalar(f'Train Loss Fold {fold}', current_loss/len(train_dataloader), global_step=epoch)
        # tb.flush()
       
        # validation
        print(f'Validating epoch {epoch}')
        val_loss = 0
        
        with torch.no_grad():
            for batch in tqdm(val_dataloader, position=0, leave=True):
    
                # get inputs
                images, labels = batch['input'], batch['gt']
                images, labels = images.to(device), labels.to(device)
    
                # generate outpuds 
                network.eval()
                preds = network(images)
                
                loss = mt_losses.dice_loss(preds, labels)
                val_loss += loss.item()
        
        mean_val_loss = val_loss/len(val_dataloader)
        
        print('mean validation loss', mean_val_loss)
        print('best loss', best_loss)
        
        if epoch % 5 == 0:
            if mean_val_loss < best_loss:
                best_loss = mean_val_loss
                save_path = f'./model-fold-{fold}.pth'
                torch.save(network.state_dict(), save_path)
            else:
                early_stopping_counter += 1
            
        tb.add_scalar(f'Validation Loss Fold {fold}', mean_val_loss, global_step=epoch)
        
        if early_stopping_counter == 25:
            # save_path = f'./model-fold-{fold}.pth'
            # torch.save(network.state_dict(), save_path)
            print('Early stopping, Model saved')
            break
        
        print(f'Early stopping counter: {early_stopping_counter}')
            
def testing_loop(network, test_dataloader, results, device, fold):
    print('Start testing')
    
    # total score
    total_score, total = 0, 0

    # evaluation for current fold
    with torch.no_grad():
        for batch in tqdm(test_dataloader, position=0, leave=True):

            # get inputs
            images, labels = batch['input'], batch['gt']
            images, labels = images.to(device), labels.to(device)

            # generate outpuds 
            network.eval()
            pred = network(images)

            # calculate dice coefficient 
            dsc = -1 * mt_losses.dice_loss(pred, labels)
            print('dice score:', dsc.item())
            total_score += dsc.item()
            total += 1

        # print dice coefficient score 
        print(f'Dice score for fold {fold}:', total_score/total)
        tb.add_scalar(f'Dice Score Test Fold {fold}', total_score/total)
        print('total score:', total_score)
        print('total preds:', total)
        print('------------------------------')
        results[fold] = total_score/total
    
    return results

def train(config):
    # get data
    dirlist = io.main(config['dataset']['data_directory'])
    filename_pairs = make_filename_pairs(dirlist, liver_path='/Liver.nii.gz')
    
    # device
    device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
    print('device:', device)
    
    # define transformations
    train_transforms = torchvision.transforms.Compose([
            # mt_transforms.RandomRotation(config['train_dataloader']['data_aug']['rot_degree']), # random rotation
            # mt_transforms.RandomAffine(0, translate=config['train_dataloader']['data_aug']['transl_range']), # shift
            # mt_transforms.RandomAffine(0, shear=config['train_dataloader']['data_aug']['shear_range']), # shear
            # mt_transforms.AdditiveGaussianNoise(mean=config['train_dataloader']['data_aug']['gauss_params']['mean'] ,std=config['train_dataloader']['data_aug']['gauss_params']['std']), # noise 
            # mt_adap.HistogramClipping(),
            # mt_adap.RangeNorm(),
            mt_transforms.ToTensor(),
        ])
    test_val_transforms = torchvision.transforms.Compose([
                                                        # mt_adap.HistogramClipping(),
                                                        # mt_adap.RangeNorm(),
                                                        mt_transforms.ToTensor()])
    
    # fold results
    results = {}
    
    # set fixed random number seed so that pseudo random number  
    # initializers will be initialized using the same initialization token.    
    # define cross validatior
    kfold = KFold(n_splits=config['model_params']['k_folds'], 
                  shuffle=True, random_state=config['model_params']['random_state'])
    
    # start print
    print('------------------------------')
    
    for fold, (train_index, test_index) in enumerate(kfold.split(filename_pairs)):
        print(f'FOLD {fold}')
        print('------------------------------')
        
        # get train and test filenames
        train_files, test_files = sub_filename_pairs(filename_pairs, train_index, test_index)
        
        # get train and validation filenames
        train_files, val_files = train_test_split(train_files, test_size=0.1, 
                                                  random_state=config['model_params']['random_state'])
        
        # define datasets
        print('start preparing datasets')
        train_dataset = MultiChannelMRI2DSegmentationDataset(train_files,
                                                         transform=train_transforms)
#                                                           slice_filter_fn=slice_filtering)
                                                         # p=0.8)
        
        val_dataset = MultiChannelMRI2DSegmentationDataset(val_files, 
                                                       transform=test_val_transforms)
#                                                         slice_filter_fn=slice_filtering)
        
        test_dataset = MultiChannelMRI2DSegmentationDataset(test_files,
                                                        transform=test_val_transforms)
#                                                         slice_filter_fn=slice_filtering)
        
        # define dataloaders
        print('start preparing dataloaders')
        train_dataloader = DataLoader(train_dataset, 
                                      batch_size=config['train_dataloader']['batch_size'],
                                      shuffle=config['train_dataloader']['shuffle'],
                                      collate_fn=mt_datasets.mt_collate)
        
        val_dataloader = DataLoader(val_dataset, 
                                    batch_size=config['test_dataloader']['batch_size'],
                                    collate_fn=mt_datasets.mt_collate)
        
        test_dataloader = DataLoader(test_dataset, 
                                     batch_size=config['test_dataloader']['batch_size'],
                                     collate_fn=mt_datasets.mt_collate)
        
        # init the neural network
        if config['model_params']['architecture'] == 'unet':
            network = MultiChannelUnet()
        elif config['model_params']['architecture'] == 'segnet':
            pass #TODO: add segnet model
        network.apply(reset_weights)
        network.to(device)
        
        # initialize optimizer
        if config['train_params']['optimizer'] == 'adam':
            optimizer = optim.Adam(network.parameters(), lr=config['train_params']['lr'])
        
        # train
        print('start training loop')
        training_loop(network, train_dataloader, val_dataloader,
                      config['train_params']['num_epochs'], 
                      device, optimizer, fold)

        # get best model        
        PATH = f'./model-fold-{fold}.pth'
        testnetwork = MultiChannelUnet()
        testnetwork.load_state_dict(torch.load(PATH))
        testnetwork.to(device)
        # test
        results = testing_loop(testnetwork, test_dataloader, 
                               results, device, fold)
        
        break 
        
    # print fold results
    k_folds = config['model_params']['k_folds']
    print(f'K-FOLD CROSS VALIDATION RESULTS FOR {k_folds} FOLDS')  
    print('------------------------------')
    sum_res = 0
    for key, value in results.items():
        print(f'Fold {key}: {value}')
        # tb.add_text(f'Fold {key}: {value}')
        sum_res += value
    print('Average:', sum_res/len(results.items()))
    
if __name__ == "__main__":
    
    CONFIG_PATH = '/home/dtank/scratch/AutoLiverSeg/src/config.yaml'
    config = load_config(CONFIG_PATH)
    
    # tensorboard logs setup
    comment = 'Liver1 Multichannel b=700 & b=10, no augmentations, no slice filter'
    tb = SummaryWriter(comment=comment)
    
    train(config)
    
    tb.close()
```
